## Big O example

### O(1) Constant

```python
def func_constant(values):
    """
    Print first item from the list
    """
    print(values[0])
```

```python
lst = [1,2,3,4]
func_constant(lst)
```

```python
%timeit func_constant(lst)
```

### O(n) Linear

```python
def func_linear(values):
    for val in values:
        print(val)
```

```python
func_linear(lst)
```

```python
%timeit func_linear(lst)
```

#### O(n)

```python
def matcher(lst,match):
    
    for item in lst:
        if item == match:
            return True
        
    return False
```

```python
lst = [1,2,3,4,5,6,7,8,9]
```
